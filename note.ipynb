```python
import numpy as np
import matplotlib.pyplot as plt

10 * np.log10((0.5 / np.sqrt(2))**2 / 50 / 0.001)

p_df = np.linspace(0.4, 0.6, 11)
p_lv = np.sqrt(2 * (1 - np.cos(2 * np.pi * p_df * 1))) / (1 * np.pi) * 0.783

plt.figure()
plt.plot(p_df, 10 * np.log10((p_lv / np.sqrt(2))**2 / 50 / 0.001))
plt.ylim(3.5, 4.0)
plt.grid()
plt.show()
```
